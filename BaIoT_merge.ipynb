```python
import pandas as pd
import pathlib
```

```python
baiot_path = pathlib.Path("data/BaIoT/")
```

```python
benign_df = pd.concat((pd.read_csv(f).apply(pd.to_numeric, downcast="float").assign(traffic_type="benign") for f in baiot_path.glob("*/benign*")), ignore_index=True)
benign_df.to_parquet("data/BaIoT/benign.parquet")
benign_df
```

```python
mirai_df = pd.concat((pd.read_csv(f).apply(pd.to_numeric, downcast="float").assign(traffic_type=f"mirai_{f.stem}") for f in baiot_path.glob("*/mirai*/*")), ignore_index=True)
mirai_df.to_parquet("data/BaIoT/mirai.parquet")
mirai_df
```

```python
bl_df = pd.concat((pd.read_csv(f).apply(pd.to_numeric, downcast="float").assign(traffic_type=f"bl_{f.stem}") for f in baiot_path.glob("*/gafgyt*/*")), ignore_index=True)
bl_df.to_parquet("data/BaIoT/bashlite.parquet")
bl_df
```

```python
combo_df = pd.concat([benign_df, mirai_df, bl_df], ignore_index=True)
combo_df.to_parquet("data/BaIoT/combined.parquet")
combo_df
```

```python
combo_df = pd.read_parquet("data/BaIoT/combined.parquet")
sample_df = combo_df.sample(frac=0.2, ignore_index=True)
sample_df.to_parquet("data/BaIoT/combined_sample.parquet")
sample_df
```
