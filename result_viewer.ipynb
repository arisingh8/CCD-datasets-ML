```python
import pickle
import pathlib

from sklearn.metrics import RocCurveDisplay

import matplotlib.pyplot as plt

#import shap

import pandas as pd
import numpy as np
```

```python
current_path = pathlib.Path(".")
```

```python
def array_round(arr):
    return list(map('{:.3f}'.format, arr))
```

```python
for score in current_path.rglob("*no_cv_binary.pickle"):
    print(score.stem)
    with open(score, "rb") as infile:
        results = pickle.load(infile)

    print("fs results -----------------------")
    print(f"number of features: {results['fs_scores'][0]['n_features']}")
    print(f"feature list: {results['fs_scores'][0]['feature_names']}")

    for result in results['no_fs_scores']:
        print(f"model: {result['model']}")
        print(f"accuracy: {result['accuracy']:.3f}")
        print(f"precision: {array_round(result['precision'])}")
        print(f"recall: {array_round(result['recall'])}")
        print(f"f1-score: {array_round(result['f1_score'])}")
        print(f"support: {list(result['support'])}")
        print(f"roc_auc_score: {result['roc_auc_score']:.6f}")
        print(f"n_features: {result['n_features']}")
        
        shap.summary_plot(shap_values=result['shap_values'],
                          feature_names=result['feature_names'])
        
        fig, ax = plt.subplots()

        print("roc curve: ")
        curve = RocCurveDisplay(fpr=result['fpr'], tpr=result['tpr'], roc_auc=result['roc_auc_score'])
        curve.plot(ax=ax, name=result['model'])

        print("---------")
```

```python
for score in current_path.rglob("*no_cv_multiclass.pickle"):
    print(score.stem)
    with open(score, "rb") as infile:
        results = pickle.load(infile)

    print("fs results -----------------------")
    print(f"number of features: {results['fs_scores'][0]['n_features']}")
    print(f"feature list: {results['fs_scores'][0]['feature_names']}")

    result_dfs = []
    for result in results['no_fs_scores']:
        print(f"model: {result['model']}")
        print(f"accuracy: {result['accuracy']:.3f}")
        print(f"precision: {array_round(result['precision'])}")
        print(f"recall: {array_round(result['recall'])}")
        print(f"f1-score: {array_round(result['f1_score'])}")
        print(f"support: {list(result['support'])}")
        print(f"roc_auc_score: {result['roc_auc_score']:.4f}")

        result_df = pd.DataFrame(data={
            "precision": np.array(array_round(result['precision'])),
            "recall": np.array(array_round(result['recall'])),
            "f1-score": np.array(array_round(result['f1_score'])),
        })
        print(result_df)
        result_dfs.append(result_df)
        
        shap.summary_plot(shap_values=result['shap_values'],
                          feature_names=result['feature_names'])

        print("---------")
    combo_results = pd.concat(result_dfs)
    combo_results.to_csv(f"results/{score.stem}.csv")
```

```python
score_path = "dohbrw_results_blb_no_cv_binary.pickle"
score_dict = {
    "Model": ["RCNN", "XCNN", "NB", "RF", "LR", "SVM", "KNN"],
    "Precision": [0.985*100, 1*100],
    "Recall": [0.985*100, 1*100],
    "F1-score": [0.99*100, 1*100]
}

with open(score_path, "rb") as infile:
    results = pickle.load(infile)

print("fs results -----------------------")
print(f"number of features: {results['fs_scores'][0]['n_features']}")
print(f"feature list: {results['fs_scores'][0]['feature_names']}")

result_dfs = []
for result in results['fs_scores']:
    score_dict["Precision"].append(np.mean(result['precision'])*100)
    score_dict["Recall"].append(np.mean(result['recall'])*100)
    score_dict["F1-score"].append(np.mean(result['f1_score'])*100)
```

```python
df = pd.DataFrame(data=score_dict).sort_values(by=['F1-score'])

x = np.arange(len(df))  # the label locations
width = 0.29  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for col in df[["Precision", "Recall", "F1-score"]]:
    offset = width * multiplier
    rects = ax.bar(x + offset, df[col], width, label=col)
    ax.bar_label(rects, fmt='%2.1f', fontsize=10)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage (%)', fontsize=14)
ax.set_title('Binary Performance on the DoHBrw Dataset', fontsize=14)
ax.set_xticks(x + width, df['Model'], fontsize=12)
ax.legend(loc='upper left', fontsize=12)
ax.tick_params(axis='y', labelsize=12)

#fig.set_tight_layout(True)
fig.set_figwidth(10)
```

```python
for score in current_path.rglob("*no_cv_binary.pickle"):
    print(score.stem)
    with open(score, "rb") as infile:
        results = pickle.load(infile)
    print("GA evolution: ")
    logbook = results['logbook']
    gafig, ax1 = plt.subplots()
    gafig.set_dpi(300)

    ax1.set_xlabel('Number of generations')
    ax1.set_ylabel('ROC-AUC score')
    ax1.plot(logbook.chapters['perf'].select("gen"), logbook.chapters['perf'].select("max"), label="Max AUC")

    ax2 = ax1.twinx()
    ax2.set_ylabel('Number of features')
    ax2.plot(logbook.chapters['n_features'].select("gen"), 
             logbook.chapters['n_features'].select("min"), 
             label="Min features",
             color="red")

    gafig.suptitle(f'BLB-GAFS Evolution on BaIoT')
    gafig.legend()
    gafig.tight_layout()

    plt.show()
```
