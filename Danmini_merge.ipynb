```python
import pandas as pd
import pathlib
from collections import defaultdict
```

```python
danmini_path = pathlib.Path("data/BaIoT/Danmini_Doorbell/")
```

```python
benign_df = pd.read_csv(danmini_path / "benign_traffic.csv").apply(pd.to_numeric, downcast="float").assign(traffic_type="benign")
benign_df
```

```python
mirai_df = pd.concat((pd.read_csv(f).apply(pd.to_numeric, downcast="float").assign(traffic_type=f"mirai_{f.stem}") for f in danmini_path.glob("mirai*/*")), ignore_index=True)
mirai_df
```

```python
bl_df = pd.concat((pd.read_csv(f).apply(pd.to_numeric, downcast="float").assign(traffic_type=f"bl_{f.stem}") for f in danmini_path.glob("gafgyt*/*")), ignore_index=True)
bl_df
```

```python
combo_df = pd.concat([benign_df, mirai_df, bl_df], ignore_index=True)
combo_df['traffic_type'] = combo_df['traffic_type'].astype("category")
combo_df.to_parquet("data/BaIoT/danmini_multiclass.parquet")
combo_df
```

```python
map_dict = defaultdict(lambda: "malicious")
map_dict['benign'] = 'benign'
combo_df['traffic_type'] = combo_df['traffic_type'].map(map_dict)
combo_df.to_parquet("data/BaIoT/danmini_binary.parquet")
combo_df
```
