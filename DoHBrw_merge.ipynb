```python
import pandas as pd
import pathlib
```

```python
doh_path = pathlib.Path("data/DoHBrw/")
```

```python
benign_df = pd.concat((pd.read_csv(f).assign(traffic_type="benign") for f in (doh_path / "benign").rglob("*.csv")), ignore_index=True).dropna().reset_index(drop=True)
benign_df
```

```python
malicious_df = pd.concat((pd.read_csv(f).assign(traffic_type=f.parent.name) for f in (doh_path / "malicious").rglob("*.csv")), ignore_index=True).dropna().reset_index(drop=True)
malicious_df
```

```python
combo_df = pd.concat([benign_df, malicious_df], ignore_index=True)
combo_df["SourceIP"] = combo_df["SourceIP"].astype("category").cat.codes
combo_df["DestinationIP"] = combo_df["DestinationIP"].astype("category").cat.codes
combo_df["SourcePort"] = combo_df["SourcePort"].astype("category").cat.codes
combo_df["DestinationPort"] = combo_df["DestinationPort"].astype("category").cat.codes
combo_df["DoH"] = combo_df["DoH"].astype("category").cat.codes
combo_df["traffic_type"] = combo_df["traffic_type"].astype("category")
combo_df.drop(["TimeStamp"], axis=1, inplace=True)
combo_df
```

```python
combo_df.to_parquet(doh_path / "multiclass.parquet")
```

```python
combo_df['traffic_type'] = combo_df['traffic_type'].map({
    'benign': 'benign',
    'dns2tcp': 'malicious',
    'dnscat2': 'malicious',
    'iodine': 'malicious'
})
```

```python
combo_df.to_parquet(doh_path / "binary.parquet")
```
